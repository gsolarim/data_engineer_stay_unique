## DataFrame Properties

### 1. Cargar los Datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
file_path = './data_rdv/Properties.csv'
df_properties = pd.read_csv(file_path, sep=',')

# Ver las primeras filas del DataFrame
print(df_properties.head())
```

### 2. Descripción General del DataFrame

```python
# Información básica del DataFrame
print(df_properties.info())

# Descripción estadística de los datos numéricos
print(df_properties.describe())
```

### 3. Comprobación de Valores Faltantes

```python
# Comprobar valores faltantes
missing_values = df_properties.isnull().sum()
print(missing_values[missing_values > 0])
```

### 4. Análisis de distribuciones

```python
# Histograma de 'Capacity'
plt.figure(figsize=(10, 6))
sns.histplot(df_properties['Capacity'], bins=10, kde=True)
plt.title('Distribución de Capacity')
plt.xlabel('Capacity')
plt.ylabel('Frecuencia')
plt.show()

# Histograma de 'Square'
plt.figure(figsize=(10, 6))
sns.histplot(df_properties['Square'], bins=10, kde=True)
plt.title('Distribución de Square')
plt.xlabel('Square')
plt.ylabel('Frecuencia')
plt.show()
```

### 5. Visualización de relaciones entre variables

```python
# Boxplot de Capacity por PropertyType
plt.figure(figsize=(12, 6))
sns.boxplot(x='PropertyType', y='Capacity', data=df_properties)
plt.title('Boxplot de Capacity por PropertyType')
plt.xlabel('PropertyType')
plt.ylabel('Capacity')
plt.xticks(rotation=45)
plt.show()

# Boxplot de NumBedrooms por PropertyType
plt.figure(figsize=(12, 6))
sns.boxplot(x='PropertyType', y='NumBedrooms', data=df_properties)
plt.title('Boxplot de NumBedrooms por PropertyType')
plt.xlabel('PropertyType')
plt.ylabel('NumBedrooms')
plt.xticks(rotation=45)
plt.show()
```

### 6. Transformación del campo PropertyType

```python
df_properties['PropertyType'] = df_properties['PropertyType'].replace({'Apa': 'Apartment'})

# Verificación de la transformación
print("\nValores únicos en PropertyType después de la transformación:")
print(df_properties['PropertyType'].unique())
```

### 7. Guardar archivo

```python
import os

# Definir la ruta de la carpeta y crearla si no existe
output_dir = 'data_ddv'
os.makedirs(output_dir, exist_ok=True)

# Guardar el DataFrame como CSV en la carpeta especificada
output_path = os.path.join(output_dir, 'Properties.csv')
df_properties.to_csv(output_path, index=False)

print(f"Archivo guardado en: {output_path}")
```
